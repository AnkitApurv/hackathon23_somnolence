```python
import pandas
```

```python
df: pandas.DataFrame = pandas.read_csv(
    filepath_or_buffer = '../../../raw/third_party/oecd/Unemployment_Rate_OECD.csv',
    header = 0
)
```

```python
df.columns
```

```python
for column in ['INDICATOR', 'SUBJECT', 'MEASURE', 'FREQUENCY', 'Flag Codes']:
    print(f'{column}: {df[column].unique()}')
```

```python
df: pandas.DataFrame = df[df['FREQUENCY'].isin(['M'])]
```

```python
df: pandas.DataFrame = df.drop(columns = ['INDICATOR', 'MEASURE', 'FREQUENCY'], inplace = False)
```

Flag Codes:

E -> Denotes data for groups of countries.

```python
g7: dict[str, str] = {
    'canada': 'CAN', 
    'france': 'FRA', 
    'germany': 'DEU', 
    'italy': 'ITA', 
    'japan': 'JPN', 
    'uk': 'GBR', 
    'usa': 'USA',
}

datasets: dict[str, pandas.DataFrame] = {}
```

```python
for country_name, country_code in g7.items():
    country_df = df[df['LOCATION'].isin([country_code])]
    country_df = country_df.drop(columns = ['LOCATION'], inplace = False)
    datasets[country_name] = country_df
```

```python
save_path = '../../../intermediate/'

for country_name, df in datasets.items():
    df.to_csv(
        path_or_buf = f'{save_path}{country_name}/unemployment_rate.csv', index = False
    )
```
